```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

nltk.download('punkt')
nltk.download('stopwords')
```

```python
file_BOW = {}
for file in file_text:
    text = file_text[file]
    # Tokenize the text
    tokens = word_tokenize(text)

    # Get English stopwords
    stop_words = set(stopwords.words('english'))

    # Remove stopwords
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

    file_BOW[file] = filtered_tokens
```

```python
file_BOW
```

```python
import numpy as np
```

```python
def doc_word(word):
    count = 0
    for file in file_BOW:
        if word in file_BOW[file]:
            count+=1
    return count

def word_count(word,file):
    count = 0
    for w in file_BOW[file]:
        if word == w:
            count+=1
    return count
            
tf_idf = {'Rain Main.txt':{},'Enemy at the gates.txt':{},
          'Fight Club.txt':{},'John Wick.txt':{},'Social Network.txt':{}}
for file in file_BOW:
    for word in file_BOW[file]:
        tf_idf[file][word] = np.log(5/doc_word(word))*word_count(word,file)/len(file_BOW[file])
```

```python
tf_idf
```

```python
def search_word(word):
    files = {'Rain Main.txt':0,'Enemy at the gates.txt':0,
          'Fight Club.txt':0,'John Wick.txt':0,'Social Network.txt':0}
    for file in tf_idf:
        for word in tf_idf[file]:
            files[file] = tf_idf[file][word]
    sorted_files = dict(sorted(files.items(), key=lambda x: x[1],reverse=True))
    return sorted_files
```

```python
def find_all_indices(item, lst):
    return [i+1 for i, elem in enumerate(lst) if elem == item]
low = text.split()
list1 = []
next_words = {}
for i in range(len(low)-1):
    next_words[low[i]] = [low[i] for i in find_all_indices(low[i],low)]
```

```python
def highest_count(l):
    temp = {}
    for word in l:
        if word not in temp:
            temp[word] = 1
        else:
            temp[word]+=1
    return sorted(temp.items(), key=lambda x: x[1],reverse=True)[0][0]

def find_next(word):
    return highest_count(next_words[word])
```
