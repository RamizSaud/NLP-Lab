```python
recipes = {"Dish":[],"Recipe":[]}
for recipe in text.split('\n'):
    recipes["Dish"].append(recipe.split(':')[0].split('.')[1].strip())
    recipes["Recipe"].append(recipe.split(':')[1])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(recipes)
```

```python
df
```

```python
df["Recipe"][0].strip()
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sen = df["Recipe"][0].strip()
sentences = sent_tokenize(sen)
print("Sentence Tokenization:", sentences)
```

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
sentence = sentences[0]
words = word_tokenize(sentence)

pos_tags = pos_tag(words)
print("POS Tagging:", pos_tags)
```

```python
final = {}
def fun(x):
    data = {"Amount":[],"Measure Type":[],"Ingredient":[],"Instruction":[],"Step":[]}
    measure_types = ['cups','cup','tsp','pinch','tbsp','minutes','lb','oz','syrup','pan','spoonfuls']
    sentences = sent_tokenize(x[1])
    for i in range(len(sentences)):
        df2 = {"Amount":[],"Measure Type":[],"Ingredient":[],"Instruction":[],"Step":[]}
        df2["Step"] = sentences[i]
        words = word_tokenize(sentences[i])
        pos_tags = pos_tag(words)
        for w,t in pos_tags:
            if t=='NNP':
                df2["Instruction"].append(w)
            elif t=='CD':
                df2["Amount"].append(w)
            elif t=='NNS' or t=='NN':
                if w in measure_types:
                    df2["Measure Type"].append(w)
                else:
                    df2["Ingredient"].append(w)
        data["Amount"].append(df2["Amount"])
        data["Measure Type"].append(df2["Measure Type"])
        data["Ingredient"].append(df2["Ingredient"])
        data["Instruction"].append(df2["Instruction"])
        data["Step"].append(df2["Step"])
    final[x[0]] = data
```

```python
pd.DataFrame(data)
```

```python
df.apply(fun,axis=1)
```

```python
final
```

```python
for i in final:
    print(i)
    display(pd.DataFrame(final[i]))
    print("\n"*3)
```
